### matplotlib包：
一个风格上类似于matlab画图的包，matplotlib对于图像区域的定义见文档地址：[matplotlib文档](https://matplotlib.org/tutorials/introductory/usage.html#sphx-glr-tutorials-introductory-usage-py)。其中，figure是整个画图区域，subplot或axes是一个差不多等价的概念（figure里的子图）。 <br/>
（1）plt.subplot()返回的是单个子图（subplot或ax）； <br/>
（2）plt.subplots()返回的是图和所有子图，这意味着同时建立图及其若干个子图； <br/>
（3）figure.subplots()返回的是figure里建立的所有子图。 <br/>
（4）matplotlib的编程风格是object-oriented API，所有的区域概念都有一个对象，而这些对象也有相应的属性和方法。例如，可以用ax和figure对象的方法去控制图像相关区域的性质。

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

# 造一个数据
x = np.linspace(-np.pi, np.pi, 20)
y1 = np.sin(x)
y2 = np.cos(x)

# 简单地画图(整体编程风格类似于matlab，只是都带着前缀plt.)
fig = plt.figure(figsize=(8, 10))

ax = plt.subplot(2, 1, 1)
plt.grid()
plt.title('y1 = sin(x)')
plt.xlabel('x')
plt.ylabel("y1")
plt.plot(x, y1, 'r-', label="sin(x)")
plt.legend()

plt.subplot(2, 1, 2)
plt.grid()
plt.title('y2 = cos(x)')
plt.xlabel('x')
plt.ylabel("y2")
plt.plot(x, y2, 'bo', label="cos(x)")
plt.legend()

plt.show()
```

```python
# 用subplots()这样的形式画图
### 同时建立图和子图
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5), sharey=True)
fig.suptitle('Using plt.subplots() to Create Plots')

ax1.plot(x, y1, 'g-x', label='sin(x)')
ax1.legend()
ax1.set_title('y1 = sinx(x)')
ax1.set_xlabel('x')
ax1.set_ylabel('y1')
ax1.grid()

ax2.plot(x, y2, 'rx', label='cos(x)')
ax2.legend()
ax2.set_title('y2 = cos(x)')
ax2.set_xlabel('x')
ax2.set_ylabel('y2')
ax2.grid()


### 在图的基础上建立子图
fig = plt.figure(figsize=(15, 5))
fig.suptitle('Using fig.subplots() to Create Plots')

ax1, ax2 = fig.subplots(1, 2, sharey=True)

ax1.scatter(y1, y2, marker='o')
ax1.set_title('y1/y2 scatter plot')
ax1.set_xlabel('y1')
ax1.set_ylabel('y2')
ax1.grid()

ax2.plot(x, y2, 'r-x', label='cos(x)')
ax2.legend()
ax2.set_title('y2 = cos(x)')
ax2.set_xlabel('x')
ax2.set_ylabel('y2')
ax2.grid()

##### 在图基础上建立子图，还可以用figure.add_subplot()一个个地画
fig = plt.figure(figsize=(10, 15))
fig.suptitle('Using fig.add_subplot() to Create Plots')

ax1 = fig.add_subplot(2, 1, 1)
ax1.scatter(y2, y1, marker='H')
ax1.set_title('y2/y1 scatter plot')
ax1.set_xlabel('y2')
ax1.set_ylabel('y1')
ax1.grid()

ax2 = fig.add_subplot(2, 1, 2)
ax2.plot(x, y2, 'y-*', label='cos(x)')
ax2.legend()
ax2.set_title('y2 = cos(x)')
ax2.set_xlabel('x')
ax2.set_ylabel('y2')
ax2.grid()
```

```python
# 运用ax和fig的方法，来控制图像的性质
fig = plt.figure(figsize=(8, 10))
fig.suptitle('Using Methods of ax/fig to Control Plot')

xticks = np.linspace(-np.pi, np.pi, 5)
xticklabels = ["$-\pi$", "$-\\frac{\pi}{2}$", "0", "$\\frac{\pi}{2}$",  "$\pi$"]

ax1 = fig.add_subplot(2, 1, 1)
ax1.plot(x, y1, 'r-')
ax1.set_xticks(xticks)
ax1.set_xticklabels(xticklabels)
ax1.set_title('y1=sin(x)')
ax1.set_xlabel('x')
ax1.set_ylabel('y1')
ax1.grid()

```
